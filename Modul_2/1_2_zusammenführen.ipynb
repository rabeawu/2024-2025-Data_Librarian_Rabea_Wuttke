```python
import pandas as pd
import json
```

```python
searchs = range(1,8)
dataframes = []
for search in searchs:
    filename = f"search{search}.json"

    with open(filename, "r", encoding="utf-8") as f:
        data = json.load(f)
        df = pd.DataFrame(data)
        df = df.loc[df.type == "Publication"]
        df = df[["label", "author", "year", "editor", "pages", "language", "type"]]
        dataframes.append(df)

df_1, df_2, df_3, df_4, df_5, df_6, df_7 = dataframes #Ki-Tipp - nochmal pr√ºfen
```

```python
df_1["search term"] = "Douglas Adams"
df_2["search term"] = "Andy Warhol"
df_3["search term"] = "Berthold Brecht"
df_4["search term"] = "Michael Ende"
df_5["search term"] = "Sherlock"
df_6["search term"] = "Forschungsdatenmanagement"
df_7["search term"] = "Python lernen"
```

```python
df_all = pd.concat([df_1, df_2, df_3, df_4, df_5, df_6, df_7])
```

```python
df_all.to_json("search_all.json", orient="records")
```
