```python
import pandas as pd
import urllib.request
```

```python
url = "https://data.seattle.gov/resource/tmmm-ytt6.csv?materialtype=MICROFORM&$limit=10000"
microf_file = "microffile.csv"
urllib.request.urlretrieve(url, microf_file)
```

```python
microf_df = pd.read_csv(microf_file)
microf_df
```

```python
year = microf_df["checkoutyear"].value_counts().sort_index()
year
```

```python
import matplotlib.pyplot as plt

year.plot(color='turquoise')
ax = plt.gca()
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
plt.xlim([2005, 2023])
plt.ylabel('Ausleihzahlen')
plt.title('AusleihhÃ¤ufigkeit von Mikroformen 2005-2023')
```
