```python
import pandas as pd
import matplotlib.pyplot as plt
import urllib.request
```

```python
url_raw = "https://data.seattle.gov/resource/tmmm-ytt6.csv?$where=checkoutyear=20"

checkouts14_file = "checkouts14_file.csv"
checkouts15_file = "checkouts15_file.csv"
checkouts16_file = "checkouts16_file.csv"
checkouts17_file = "checkouts17_file.csv"
checkouts18_file = "checkouts18_file.csv"
checkouts19_file = "checkouts19_file.csv"
checkouts20_file = "checkouts20_file.csv"
checkouts21_file = "checkouts21_file.csv"
checkouts22_file = "checkouts22_file.csv"
checkouts23_file = "checkouts23_file.csv"
checkouts24_file = "checkouts24_file.csv"

years = list(range(14, 25))
years_str = [str(year) for year in years]
```

```python
for year in years_str:
    urllib.request.urlretrieve(url_raw + year, "checkouts" + year + "_file.csv")
```

```python
checkouts14_df = pd.read_csv(checkouts14_file)
checkouts15_df = pd.read_csv(checkouts15_file)
checkouts16_df = pd.read_csv(checkouts16_file)
checkouts17_df = pd.read_csv(checkouts17_file)
checkouts18_df = pd.read_csv(checkouts18_file)
checkouts19_df = pd.read_csv(checkouts19_file)
checkouts20_df = pd.read_csv(checkouts20_file)
checkouts21_df = pd.read_csv(checkouts21_file)
checkouts22_df = pd.read_csv(checkouts22_file)
checkouts23_df = pd.read_csv(checkouts23_file)
checkouts24_df = pd.read_csv(checkouts24_file)
```

```python
bestseller14 = checkouts14_df[checkouts14_df["checkouts"] == (checkouts14_df["checkouts"].max())]
bestseller15 = checkouts15_df[checkouts15_df["checkouts"] == (checkouts15_df["checkouts"].max())]
bestseller16 = checkouts16_df[checkouts16_df["checkouts"] == (checkouts16_df["checkouts"].max())]
bestseller17 = checkouts17_df[checkouts17_df["checkouts"] == (checkouts17_df["checkouts"].max())]
bestseller18 = checkouts18_df[checkouts18_df["checkouts"] == (checkouts18_df["checkouts"].max())]
bestseller19 = checkouts19_df[checkouts19_df["checkouts"] == (checkouts19_df["checkouts"].max())]
bestseller20 = checkouts20_df[checkouts20_df["checkouts"] == (checkouts20_df["checkouts"].max())]
bestseller21 = checkouts21_df[checkouts21_df["checkouts"] == (checkouts21_df["checkouts"].max())]
bestseller22 = checkouts22_df[checkouts22_df["checkouts"] == (checkouts22_df["checkouts"].max())]
bestseller23 = checkouts23_df[checkouts23_df["checkouts"] == (checkouts23_df["checkouts"].max())]
bestseller24 = checkouts24_df[checkouts24_df["checkouts"] == (checkouts24_df["checkouts"].max())]
```
