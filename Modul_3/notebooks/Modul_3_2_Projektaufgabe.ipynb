## Beschreibung des Datensatz

Es handelt sich um einen Datensatz in dem Ausleihdaten der NutzerInnen der San Francisco Public Library abgebildet sind. Die Grundgesamtheit bilden die aktiven NutzerInnen der Bibliothek, seit 2004 bis 2024. Der Datensatz besteht aus 450359 Stichproben, Merkmalsträger sind die NutzerInnen zu denen bis zu 11 Merkmale (Variablen) erhoben wurden.

```python
df.columns
```

```python
df['Age Range'] = df['Age Range'].fillna("keine Angaben")
```

```python
df['Notice Preference Definition'] = df['Notice Preference Definition'].fillna("keine Angaben")
```

```python
df['Home Library Definition'] = df['Home Library Definition'].fillna("keine Angaben")
```

```python
df.isna().sum()
```

```python
df_clean = df.dropna()
```

```python
df_clean.isna().sum()
```

```python
df_clean
```

```python
import numpy as np
```

```python
pd.to_numeric(df_clean['Circulation Active Year'], errors='coerce')
```

```python
df_clean['Circulation Active Month'] = pd.to_datetime(
    df_clean['Circulation Active Month'],
    errors='coerce',
    format="%b"
)
```

```python
df_clean['Circulation Active Month'] = df_clean['Circulation Active Month'].dt.month
```

```python
df_clean['Membership Duration'] = (df_clean['Circulation Active Year'] - df_clean['Year Patron Registered'])*12 + df_clean['Circulation Active Month']
```
